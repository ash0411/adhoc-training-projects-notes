1. to find all the titles of link as well as the links given in a particular page 

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Lists_of_countries_and_territories'
webdata = requests.get(url)
data = webdata.text
print(data)
```

```python
souped = BeautifulSoup(data,'html5lib')
print(souped.prettify())
```

```python
initial = souped.find('div',{'class':'mw-body'})
```

```python
a_tags= souped.findAll('a')
```

```python
links = []
for link in a_tags:
    links.append(link.get('title'))
print(links)
```

```python
links_h = []
for link in a_tags:
    links_h.append(link.get('href'))
print(links_h)
```
