```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# now creating data
x=np.arange(0,100,5)
x1=np.arange(0,100,4)
y =[ i for i in range(len(x))]
y1 =[ i for i in range(len(x1))]
```

```python
plt.xlabel("random x")
plt.ylabel("random y")
plt.xlim(0,100)
plt.ylim(0,25)
plt.plot(x,y,label='numpy data view')
plt.bar(x,y,label='numpy data view')
plt.plot(x1,y1,label='numpy data random')
plt.bar(x1,y1,label='numpy data random')
plt.legend()
plt.grid(color='g')
plt.show()
```

```python
players =["virat",'rohit','dhoni','jadeja']
runs = [1200,550,700,500]
runs1 =[1300,900,1000,800]
runs2 =[1500,1900,1200,1800]
```

```python
plt.xlabel("cricket players")
plt.ylabel("runs")
plt.ylim(400,1300)
plt.bar(players,runs,label = "performance in 2019")
#plt.bar(players,runs1,label = "performance in 2020")
plt.grid(color='g')
plt.legend()
plt.show()
```

```python
#subplot
# here subplot will give you 2 properties
# fig will be for entire figure
# ax will be for indiviual attributes
#for graph1 2019
fig,ax=plt.subplots(3)
#fig[0].ylim(0,2000)
ax[0].bar(players,runs,label="2019")
ax[0].legend()
#ax[0].set_title("performance of 2019")
#for graph 2 2020
#ax[1].ylim(0,2000)
ax[1].bar(players,runs1,color="green",label ="2020")
fig.suptitle("cricket performance class")
ax[1].legend()
#ax[2].ylim(0,2000)
ax[2].bar(players,runs2,label="2021",color='r')
ax[2].legend()
#ax[1].set_title("performance of 2020")
plt.show()
```

```python
#pie chart
#plt.xlabel("cricket players")
#plt.ylabel("runs")
plt.pie(runs2,labels=players,shadow=True,autopct ='%1.1f%%',explode=(0,0.1,0,0))#%1.1f% is a formula and % is for writing percentage
plt.legend()
plt.title("runs of players")
plt.show()
```

```python
fig,ax=plt.subplots(2)
ax[0].pie(runs2,labels=players,shadow=True,autopct ='%1.1f%%',explode=(0,0.1,0,0))#%1.1f% is a formula and % is for writing percentage
ax[0].legend()
ax[1].pie(runs1,labels=players,shadow=True,autopct ='%1.1f%%',explode=(0,0,0.2,0))#%1.1f% is a formula and % is for writing percentage
ax[1].legend()
plt.show()
```

```python
plt.xlabel("random x")
plt.ylabel("random y")
plt.xlim(0,100)
plt.ylim(0,25)
plt.scatter(x,y,label='numpy data view',marker ="^",s=100)
#plt.bar(x,y,label='numpy data view')
plt.scatter(x1,y1,label='numpy data random')
#plt.bar(x1,y1,label='numpy data random')
plt.legend()
plt.grid(color='g')
plt.show()
```
